# Part A: Chernoff Bound

Suppose $X_1, X_2, \cdots, X_n$ are i.i.d. copies of a $\mathcal{N}(0,\sigma^2)$ r.v. Then for $X = \frac{1}{n} \sum_{i=1}^n X_i$ we know that $$ \mathbb{P} [X \geq \epsilon] \leq \exp\left(\dfrac{-n\epsilon^2}{2\sigma^2}\right)  $$

Write a Python code to run Monte Carlo simulations that verify the inequality. Specifically, for a given  $\epsilon$ and $\sigma$, generate $n$ samples from the zero mean Gaussian distribution $x_1, x_2, \dots, x_n$ and check whether the sample average is more than $\epsilon$. Repeat this experiment 500 times and observe in how many experiments out of those 500 experiments, the sample average is more than $\epsilon$. This will gives us an empirical estimate of ${P}[X\geq \epsilon]$.

Take $\sigma = 0.1,~ \epsilon = 0.05$, and plot the empirical estimate as a function of $n \in \{100, 200, \dots, 1000\}$. In the same plot, include the Chernoff upper bound as a function of $n$. `[10 Marks]`

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def chernoff_upper_bound(n, epsilon, sigma):
    # Chernoff upper bound formula
    max_val = np.exp(-n * epsilon**2 / (2 * sigma**2))
    return max_val
```

```python

def monte_carlo(n, epsilon, sigma, num_exp=500):
    true_outcomes = 0

    for _ in range(num_exp):
        # Generate n samples from the zero mean Gaussian distribution
        samples = np.random.normal(loc=0, scale=sigma, size=n)
        # It will give samples x1,x2,x3......xn
        # Check if the sample average is more than epsilon
        sample_average = np.mean(samples)
        if sample_average >= epsilon:
            true_outcomes += 1

    # Empirical estimate of P[X >= epsilon]
    empirical_estimate = true_outcomes / num_exp
    return empirical_estimate
```

```python
#  Given Parameters
sigma = 0.1
epsilon = 0.05 
num_exp = 500 # No. of experiments
n_values = np.arange(1, 51)
```

```python
# Run Monte Carlo simulations and calculate empirical estimates
empirical_estimates = [monte_carlo(n, epsilon, sigma, num_exp) for n in n_values]

# Calculate Chernoff upper bounds
chernoff_bounds = [chernoff_upper_bound(n, epsilon, sigma) for n in n_values]
```

```python
# Plotting the graph
plt.figure(figsize=(14, 6))
plt.scatter(n_values, empirical_estimates, label='Empirical_Estimate', marker='o',color='red',s=100)
plt.plot(n_values, chernoff_bounds, label='Chernoff Upper Bound')
plt.xlabel('n')
plt.ylabel('P[X >= epsilon]')
plt.title('Empirical Estimates and Chernoff Upper Bound')
plt.legend()
plt.show()
```
